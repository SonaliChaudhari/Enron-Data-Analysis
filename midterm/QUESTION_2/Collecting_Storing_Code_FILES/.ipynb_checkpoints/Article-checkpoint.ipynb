```python
import os
import requests
from pprint import pprint
import json
from datetime import datetime
import glob
import time
```

```python
cur_path = os.path.dirname(os.getcwd())
print(cur_path)
path = cur_path + '/data/NYT_data'
print (path)
```

```python
key1 = os.environ.get('API_KEY2')
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?api-key="+(key1)+"&begin_date=20081126&end_date=20081231"
```

```python
i = 0
response_data = []
for i in range(48):
    try:
        response_data.append(requests.get(url).json())
        i += 1
        time.sleep(3)
    except ValueError:
        continue
```

```python
response_data
```

```python
articles_path = path + '/Article/'
for json_file in response_data:
    with open(articles_path + 'Atricle2008'+ str(datetime.now()) +'.json', 'w') as outfile:
        json.dump(json_file, outfile)
```

```python
import glob
json_file_path = articles_path +'/*'
lambda_file = lambda json_file_path : glob.glob(json_file_path)
```

```python
article_search_dict = {}
for json_file in lambda_file(json_file_path):
    with open(json_file) as json_data:
        article_search_dict.update(json.load(json_data))
    pprint(article_search_dict)
```
