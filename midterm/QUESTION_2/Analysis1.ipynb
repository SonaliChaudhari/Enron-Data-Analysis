## ALL Require Imports

```python
import json
from pprint import pprint
import glob
import os
import re
import datetime
import string
from nltk.corpus import stopwords  # Using NLTK
from collections import Counter
```

```python
#Retrieving the current system path
directory = os.path.dirname(os.getcwd()) 

# Retreiving the required path and assigning it to variable 'reqpath'
reqpath = directory + '/midterm/data/NYT_data/article_search/*'

reqpath
```

```python
Words=[]
for file in glob.glob(reqpath):
    with open(file) as f: 
        #print(f)
        data = json.load(f)
        x = data['response']['docs']
        for each in x:
            snippet=str(each['snippet'])
            words=re.findall(r'(\b[A-Za-z][a-z]{2,9}\b)', snippet)        
            Words.append(words)
Words=sum(Words,[])
      
```

```python
Words
```

## Removing Stopwords and punctuation

```python
Words=[''.join(c for c in s if c not in string.punctuation) for s in Words]
Words = [s for s in Words if s]
```

```python
len(Words)
```

```python
stop = stopwords.words('english')
stop.append('The')
newsWords = [word for word in Words if word not in stop]
```

```python
len(newsWords)
```
