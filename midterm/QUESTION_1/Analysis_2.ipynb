## Function to return only the emailids from a string

```python
def mainlist(list,dictionary):
    for item in list:
        if item != None:
            p = item.split(",")
            for i in p:
                match = re.findall(r'[\w\.-]+@[\w\.-]+', i)   
                for a in match:
                    if a not in dictionary:
                        dictionary[a] = 1
                    else:
                        dictionary[a] += 1  
    return dictionary
```

## Removing blank and unwanted spaces and entries

```python
for s in subj:
    s = s.lstrip('Re: ')
    s = s.lstrip('FW: ')
    s = s.lstrip('Re: FW: ')
    if s == ' ':
        break
    else:
        if s not in SUB:
            SUB[s]=1
        else:
            SUB[s]=+1    
```

```python
S = sorted(SUB.items(), key=lambda x: x[1], reverse=True)[:len(SUB)]
```

```python
with open('Subjects.csv', 'w') as csvfile:    # Creating a CSV file named 'Subjects' in write mode as csvfile
    fieldnames = ['Subjects','COUNTS']  #Naming the title Row of the CSV
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)    
    for k,v in S:
        writer.writerow({'Subjects':k,'COUNTS':v}) # Adding entry in the CSV
```
