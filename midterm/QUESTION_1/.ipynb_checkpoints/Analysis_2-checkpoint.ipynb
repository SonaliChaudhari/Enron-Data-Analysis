```python
def mainlist(list,dictionary):
    for item in list:
        if item != None:
            p = item.split(",")
            for i in p:
                match = re.findall(r'[\w\.-]+@[\w\.-]+', i)   
                for a in match:
                    if a not in dictionary:
                        dictionary[a] = 1
                    else:
                        dictionary[a] += 1  
    return dictionary
    
```
